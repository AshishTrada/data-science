```python
import requests
```

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome('D:\chromedriver.exe')
```

```python
products_name = []
prices=[]
actual_prices=[]
```

```python
driver.get('https://www.bewakoof.com/search/shirt/gender-men')
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
```

```python
content
```

```python
count = 1
while count<25:
    for a in soup.findAll('div', attrs={'class':'productCardDetail '}):
        name = a.find('h3')
        price = a.find('span', attrs={'class':'discountedPriceText'})
        original_price = a.find('span', attrs={'class': 'actualPriceText'})
        if (len(name.text)>0 or len(price.text)>0 or len(original_price.text)>0):
            products_name.append(name.text)
            prices.append(price.text)
            actual_prices.append(original_price.text)
    count = count+1
```

```python
df = pd.DataFrame({'Product Name': products_name, 'discounted Price':prices, 'original Price': actual_prices})
df
```

```python
df.to_csv('scraped_products.csv', index=False, encoding='utf8')
```
